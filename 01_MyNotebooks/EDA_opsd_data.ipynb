#### Correlation, Autocorrelaiton and Partial Autocorrelation

```python
import seaborn as sns

corr = df.loc[:,:'DE_radiation_diffuse_horizontal'].corr()

sns.heatmap(corr, annot=True)
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(df.AT_load_actual_entsoe_transparency, lags=48, title='Autocorrelation 48h')
```

```python
plot_acf(df.AT_load_actual_entsoe_transparency, lags=336, title='Autocorrelation 336h (2 Weeks)')
```

```python
plot_acf(df.AT_load_actual_entsoe_transparency, lags=1448, title='Autocorrelation 1488h (2 Months)')
```

```python
plot_acf(df.AT_load_actual_entsoe_transparency, lags=8760, title='Autocorrelation 8760h (1 Year)') 
```

```python
plot_pacf(df.AT_load_actual_entsoe_transparency, lags=48, title='Autocorrelation 48h')
```

```python
plot_pacf(df.AT_load_actual_entsoe_transparency, lags=336, title='Autocorrelation 336h (2 Weeks)')
```

```python
plot_pacf(df.AT_load_actual_entsoe_transparency, lags=1448, title='Autocorrelation 1488h (2 Months)')
```

```python
plot_pacf(df.AT_load_actual_entsoe_transparency, lags=8760, title='Autocorrelation 8760h (1 Year)')
```

#### Results

+ Correlation with solar radiatio is high
+ Autocorrelations for one day, month, 6 months and 1 year are high
+ Partial Autocorrleation suggests, that data which is older than one week is not correlated much.
